```python
#importing the libraries 
import pandas as pd
import numpy as np
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
```

```python
#loading the datasets 
train= pd.read_csv(r"C:\Users\jiten\Downloads\hacktrain (1).csv")
test = pd.read_csv(r"C:\Users\jiten\Downloads\hacktest.csv")
```

```python
# Removing extra unnamed columns if present
if 'Unnamed: 0' in train.columns:
    train = train.drop(columns=['Unnamed: 0'])

if 'Unnamed: 0' in test.columns:
    test = test.drop(columns=['Unnamed: 0'])
```

```python
# Separating features and labels from training data
X_train = train.drop(columns=["ID", "class"])
y_train = train["class"]
```

```python
# Droping the ID column from test data
X_test = test.drop(columns=["ID"])
```

```python
# Filling the missing values with the mean of each column
imputer = SimpleImputer(strategy="mean")
X_train_filled = imputer.fit_transform(X_train)
X_test_filled = imputer.transform(X_test)
```

```python
# Scaling the data so all values are on a similar range
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_filled)
X_test_scaled = scaler.transform(X_test_filled)
```

```python
# Using PCA to reduce the number of features and reduce noise
pca = PCA(n_components=10)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
# Converting class labels to numbers
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
```

```python
# Build and train the logistic regression model
model = LogisticRegression(max_iter=1000,solver='lbfgs')
model.fit(X_train_pca, y_train_encoded)
```

```python
# Making predictions on the test data
predictions = model.predict(X_test_pca)
```

```python
# Converting numeric predictions back to original class names
final_predictions = encoder.inverse_transform(predictions)
```

```python
# Saving the results into a CSV file
output = pd.DataFrame({
    "ID": test["ID"],
    "class": final_predictions
})
```

```python
output.to_csv("submission.csv", index=False)
```
